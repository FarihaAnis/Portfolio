```python
#Initial summary exploration
df1.info()
```

```python
#Filter all columns with string data type
str_columns = df1.select_dtypes(include=['object']).columns
```

```python
#Remove newline characters from each string column
for column in str_columns:
    df1[column] = df1[column].str.replace('\n', '')
```

```python
#Convert relevant columns into numerical

#Skill Moves Rating
df1['Skill Moves Rating'] = df1['Skill Moves Rating'].str.replace('★', '').astype(int)
```

```python
# Function to convert height strings to centimeters
def convert_height(height_str):
    # Check if the height is in feet and inches format
    if '\'' in height_str:
        # Split feet and inches
        feet, inches = height_str.split('\'')
        # Calculate total inches
        total_inches = int(feet) * 12 + int(inches.replace('"', '').strip())
        # Convert inches to centimeters
        height_cm = total_inches * 2.54
    else:
        # Remove 'cm' and convert to integer
        height_cm = int(height_str.replace('cm', '').strip())
    
    return height_cm

# Apply the conversion function to the 'Height' column
df1['Height'] = df1['Height'].apply(convert_height)
```

```python
# Function to convert weight strings to kilograms and round to 2 decimal places
def convert_weight(weight_str):
    if 'kg' in weight_str:
        # If weight is already in kilograms, remove 'kg' and convert to integer
        weight_kg = int(weight_str.replace('kg', '').strip())
    elif 'lbs' in weight_str:
        # If weight is in pounds, convert to kilograms (1 pound = 0.453592 kilograms)
        weight_lbs = float(weight_str.replace('lbs', '').strip())
        weight_kg = round(weight_lbs * 0.453592, 2)
    else:
        # Handle any other format or units as needed
        weight_kg = None
    
    return weight_kg

# Apply the conversion function to the 'Weight' column
df1['Weight'] = df1['Weight'].apply(convert_weight)
```

```python
# Function to convert contract strings to numerical values
def contract_to_years(contract_str):
    if 'On Loan' in contract_str or contract_str == 'Free':
        return 0  # Treat loans and Free as 0 years
    elif '~' in contract_str:
        start, end = contract_str.split('~')
        return int(end.strip()) - int(start.strip())
    else:
        return int(contract_str.split(' ')[0])

# Apply the function to the 'Contract' column and create a new column 'Contract_Years'
df1['Contract'] = df1['Contract'].apply(contract_to_years)
```

```python
#Market Value
df1['Value'] = df1['Value'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3')
df1['Value'] = df1['Value'].apply(pd.eval).astype(float)
```

## Handling Outliers

Handling outliers involves a systematic approach to address extreme values that may impact the interpretation of player data. In this analysis, initially, a threshold is defined based on the Interquartile Range (IQR), to identify potential outliers in attributes. These outliers are assessed for their realism within the context of professional football and the FIFA 21 simulation. Outliers that align with the expectations and constraints of real-world football are capped, ensuring that the data remains consistent with the sport's dynamics. This process maintains the realism of player attributes while mitigating the influence of extreme values on the analysis.

```python
# Define the threshold for outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Create an empty dictionary to store the total outlier counts
total_outlier_counts = {}

# Loop through numeric columns
for column in numeric_columns.columns:
    # Calculate the IQR for the current column
    Q1 = numeric_columns[column].quantile(0.25)
    Q3 = numeric_columns[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define conditions to identify outliers for the current column
    outliers = (numeric_columns[column] > (Q3 + threshold * IQR))
    
    # Count the total number of outliers for the current column
    total_outlier_count = outliers.sum()
    
    # Store the total outlier count in the dictionary
    total_outlier_counts[column] = total_outlier_count

# Print the total outlier counts for each numeric column
for column, count in total_outlier_counts.items():
    print(f"Total outlier count for {column}: {count}")
```

### Outliers Capping


```python
# Function to cap outliers using the IQR method
def cap_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Cap values above the upper bound
        df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else x)
        
        # Cap values below the lower bound
        df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else x)
    
    return df

# Define the columns for which to cap outliers
columns_to_check = ['Age', 'Contract Duration (Years)', 'Weight', 'Height']

# Cap outliers using the IQR method and update df1
df1 = cap_outliers_iqr(df1, columns_to_check)

# Visualize boxplots after capping outliers
plt.figure(figsize=(12, 8))
for i, column in enumerate(columns_to_check, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df1[column])
    plt.title(f'Capped {column}')

plt.tight_layout()
```

The decision to limit outliers in some attributes while allowing them in others is rooted in the pursuit of a realistic gaming experience that mirrors professional football. Outliers in age are capped since it's exceptionally rare for players to compete in their 40s or 50s due to physical decline. Likewise, contract durations beyond 5 years are infrequent in football, and extremely long contracts are capped to align with real-world norms. Weight is capped at a certain threshold to reflect the sport's need for a balance between strength and endurance. Height outliers are restricted because players taller than 205 cm face challenges with agility and balance. However, outliers are maintained in skill-related attributes like Reactions, Vision, Composure, Dribbling, Overall Rating (OVA), and Potential Rating to represent the diverse skill levels and growth potential found in professional football, enhancing the game's authenticity.

## Data Discretization and Binning


In the FIFA 21 player attributes analysis, data discretization and binning play a pivotal role in categorizing player ages into meaningful groups for better insights. The process begins by defining age bins, which represent distinct age ranges relevant to professional football. Corresponding labels are assigned to these bins, providing human-readable categories such as 'Youth,' 'Prime,' and 'Experienced.' Subsequently, player ages are discretized into these categories using the specified bins and labels, effectively replacing the 'Age' column with the newly created 'Age Group' categories. Finally, a bar plot is generated to visualize the player count in each age group, offering a clearer understanding of the player distribution across different stages of their football careers. This data discretization and binning approach aids in the analysis of player attributes based on age groups, facilitating insights into how different age categories impact player performance in FIFA 21.

```python
df1[['Age']].describe()
```

```python
#Define age bins
bins = [16, 25, 30, 42]

#Define the corresponding labels for bins
labels = ['Youth', 'Prime', 'Experienced']

#Discretize age into categories
df1['Age Group'] = pd.cut(df1['Age'], bins=bins, labels=labels, right=False)

#Drop the 'Age' column
df1.drop(columns=['Age'], inplace=True)
```
